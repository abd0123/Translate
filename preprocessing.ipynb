```python
ar = ['ز','ظ','و','ة','ى','ر','ؤ','ء','ئ','ط','م','ن','ت','ا','ل','ب','ي','س','ش','د','ج','ح','خ','ه','ع','غ','إ','ف','ق','ث','ص','ك','ض','ذ',' ', 'أ','آ', '~']
```

## Quran

```python
with zipfile.ZipFile('Arabic-Original.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('./')
```

```python
arabic = pd.read_csv(r'Arabic-Original.csv', names = ['Surah', 'Ayat', 'Verse'],delimiter= '|' )
english = pd.read_csv(r'Quran_English.csv')
print(arabic.shape)
print(english.shape)
```

```python
def clean(verse: str):
  res = ""
  for i in range(len(verse)):
    if verse[i] in ar:
      res+=verse[i];
  return res
```

```python
for i in range(6236):
  arabic.at[i,'Verse'] = clean(arabic.at[i,'Verse'])
  if i != 0:
    if 'بسم الله الرحمن الرحيم' in arabic.at[i, 'Verse']:
      arabic.at[i, 'Verse'] = arabic.at[i, 'Verse'].replace('بسم الله الرحمن الرحيم', '')
```

```python
arabic.at[1, 'Verse']
```

```python
for i in range(6236):
  if arabic.iloc[i]['Surah']!=english.iloc[i]['Surah']:
    print(arabic.iloc[i])
    print(english.iloc[i])
  if arabic.iloc[i]['Ayat']!=english.iloc[i]['Ayat']:
    print(arabic.iloc[i])
    print(english.iloc[i])
```

```python
columns = {'verse_ar':[], 'verse_eng':[]}
df = pd.DataFrame(columns)
```

```python
for i in range(6236):
  dict = {'verse_ar': arabic.iloc[i]['Verse'], 'verse_eng': english.iloc[i]['Verse']}
  df = df.append(dict, ignore_index=True)
```

```python
df.to_csv('en-ar.csv',index=False)
```
